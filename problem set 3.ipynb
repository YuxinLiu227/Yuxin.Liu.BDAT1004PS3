# Question 1

```python
# Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Step 2. Import the dataset
# Step 3. Assign it to a variable called users
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

```python
# Step 4. Discover what is the mean age per occupation
users[['age', 'occupation']].groupby('occupation').mean()
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
users_5 = users[['user_id', 'gender', 'occupation']].groupby(['occupation', 'gender']).count().unstack()
users_5.fillna(0, inplace=True)
users_5['Male ratio'] = users_5[('user_id', 'M')]/(users_5[('user_id', 'M')]+ users_5[('user_id', 'F')])
users_5['Male ratio'].sort_values(ascending=False)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
users[['age', 'occupation']].groupby('occupation').max()
```

```python
users[['age', 'occupation']].groupby('occupation').min()
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
users[['age', 'gender', 'occupation']].groupby([ 'occupation', 'gender']).mean()
```

```python
# Step 8. For each occupation present the percentage of women and men
users_8 = users[['user_id', 'gender', 'occupation']].groupby(['occupation', 'gender']).count().unstack()
users_8.fillna(0, inplace=True)
gender_sum = users_8[('user_id', 'M')]+ users_8[('user_id', 'F')]
users_8[('user_id', 'M')] /= gender_sum
users_8[('user_id', 'F')] /= gender_sum
users_8
```

```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from this address
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', sep=r'[ ]+')
data.Yr = data.Yr.apply(lambda x:x+2000)
data['date'] = pd.to_datetime(data[['Yr','Mo', 'Dy']].rename({'Yr': 'year', 'Mo': 'month', 'Dy': 'day'}, axis=1))
data.drop(['Yr','Mo', 'Dy'], axis=1, inplace=True)
data
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fix_year(x):
    Y, m, d = x.year-100, x.month, x.day
    return pd.to_datetime(f'{Y}-{m}-{d}')
data.date = data.date.apply(fix_year)
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
data.set_index('date', inplace=True)
data
```

```python
# Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
data.isnull().sum()
```

```python
# Step 7. Compute how many non-missing values there are in total.
data.notnull().sum()
```

```python
# Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
data.mean().mean()
```

```python
# Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1)
loc_stats.rename({0:'min', 1:'max', 2:'mean', 3:'std'}, axis=1, inplace=True)
loc_stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
day_stats.rename({0:'min', 1:'max', 2:'mean', 3:'std'}, axis=1, inplace=True)
day_stats
```

```python
# Step 11. Find the average windspeed in January for each location.
data[data.index.month==1].mean()
```

```python
# Step 12. Downsample the record to a yearly frequency for each location.
data.resample('365d').first()
```

```python
# Step 13. Downsample the record to a monthly frequency for each location.
data.resample('1m').first()
```

```python
# Step 14. Downsample the record to a weekly frequency for each location.
data.resample('1w').first()
```

```python
# Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.
data15 = data.iloc[1:].resample('7d').first().iloc[:52]
loc_stats15 = pd.concat([data15.min(), data15.max(), data15.mean(), data15.std()], axis=1)
loc_stats15.rename({0:'min', 1:'max', 2:'mean', 3:'std'}, axis=1, inplace=True)
loc_stats15
```
